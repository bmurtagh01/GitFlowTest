## 1. Most Retweeted

```python
def most_tweeted(db):
    top_10 = db.sort_values('retweetCount', ascending=False).head(10)['content'].tolist()
    print("-- Top 10 Most Retweeted --")
    print()
    for number, tweet in enumerate(top_10, 1):
        print(str(number) + ".", tweet)
        print()
```
